```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
data_source = 'artwork_data.csv'
df = pd.read_csv(
    data_source, 
    index_col='id',
    usecols=['id', 'artist', 'title', 'medium', 'year', 'acquisitionYear', 'width', 'height', 'units']
)
print(df.shape)
print(df.head())
```

```python
# Simple Default Matplotlib plot
```

```python
df.groupby('acquisitionYear').size().plot()
# df.groupby('year').size().plot()
```

```python
# Manual setup
```

```python
from matplotlib.pyplot import figure
from matplotlib import rcParams
```

```python
acquisition_year_groups = df.groupby('acquisitionYear').size()
```

```python
fig = figure(
    figsize=(14, 8),
)
fig.set_tight_layout(False)  # prevent incompatible axis warning

rcParams.update({ 'figure.autolayout': True, })
subplot = fig.add_subplot(1, 1, 1)

acquisition_year_groups.plot(
    ax=subplot, 
    rot=45,
    logy=True,
)

subplot.set_xlabel('Year of Acquisition')
subplot.set_ylabel('Number of acquired works')
subplot.locator_params(nbins=50, axis='x')
```

```python
# Plot export to file
```

```python
fig.savefig('acquisition_plot.png')
fig.savefig('acquisition_plot.svg', format='svg')
```
